```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from neural_network_2_layers import MLP_Two_Layers
```

```python
breat_cancer_data = load_breast_cancer()
```

```python
X = breat_cancer_data['data']
y = breat_cancer_data['target']

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    stratify=y,
                                                    random_state=42
                                                   )
```

```python
X.shape
```

```python
y.shape
```

```python
model = MLP_Two_Layers(
    input_size=30, hidden_size=15, output_size=1, random_state=42
)
```

```python
model.fit(X_train,y_train,batch_size=128,epochs=900,learning_rate=0.001,beta=0.9)
```

```python
y_predict = model.predict(X_test)

print(classification_report(y_test,
                            y_predict,
                            target_names=['positive','negative'],
                            digits=3
                           )
     )
```
